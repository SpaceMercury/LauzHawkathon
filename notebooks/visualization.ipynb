```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../clean_data/BRISTOR_Zegoland_all_%.csv')


# Assuming your data is loaded into a DataFrame called `df`
# Ensure the 'Date' column is in datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Create a plot for each column except 'Date'
for column in df.columns:
    if column != 'Date' :  # Skip the 'Date' column
        plt.figure(figsize=(10, 6))
        plt.plot(df['Date'], df[column], label=column, marker='o')
        plt.xlabel('Date')
        plt.ylabel(column)
        plt.title(f'{column} Over Time')
        plt.legend()
        plt.grid(True)
        plt.show()
```
